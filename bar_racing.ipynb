```python
df2 = df[['date', 'China', 'Italy', 'Spain', 'Turkey', 'US', 'United Kingdom', 'Sweden', 'Germany', 'Finland','France', 'Egypt', 'Netherlands', 'Portugal', 'Russia', 'Ukraine', 'New Zealand', 'Israel', 'Hungary', 'Croatia',]]
#tarih degiskenimizi tutup gormek istedigimiz kolonlari seciyoruz. df2 degiskenimize atiyoruz
```

```python
df2.head()
```

```python
df2.info()
```

```python
df2.set_index("date", inplace=True)
#date'r gore yani tarihe gore ulkelerin korona verilerine baktik. set index ile veriyi index olarak ayarladik.
#yani date kolonunu indeks yani 0 1 2 3 4 satiri haline getirdik.
#inplace true parametresi ile de yapilan degisiklikleri kalici hale getirdik.
```

```python
df2.head()
```

```python
df_total = df2.cumsum(axis=0)  #kumulatif veri

#cumsum: kumulatif toplamlari hesaplamak icin kullanilir. axis=0 satirdaki axia=1 sutundaki
#amacimiz her ulkenin korona verilerini her ulke bazinda kamulatif toplayip diger ulkelerle karsilastirmak
#df_total degisknine atayip ana verimizi keybetmeden bu degisken uzerinden devam ettik
#df_total bizimyeni veri cercevemiz oldu. bu veri cercevesinin her bir sutunu ilgili sutunun kumulatif toplamini icerir.
```
